## Implement a random trader

To check whether the code works, we first turn off Q-learning and the Q-learner will return random actions. 
We then plot cumulative returns from random tradings on both training and testing data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
import util as ut
import numpy as np
import QLearner as ql
import StrategyLearner as sl

# Pretty display for notebooks
%matplotlib inline
```

Calculate benchmarks:

```python
# set parameters for training the learner
symbol = "IBM"
syms=[symbol]

sd_train = dt.datetime(2007,12,31)
ed_train = dt.datetime(2009,12,31) 

train_dates = pd.date_range(sd_train, ed_train)
train_prices_all = ut.get_data(syms, train_dates)  # automatically adds SPY
train_prices = train_prices_all[syms]  # only portfolio symbols

# set parameters for testing
sd_test = dt.datetime(2009,12,31)
ed_test = dt.datetime(2011,12,31) 

dates = pd.date_range(sd_test, ed_test)
prices_all = ut.get_data(syms, dates)  # automatically adds SPY
prices = prices_all[syms]  # only portfolio symbols

# compute benchmark
sv =10000
train_cumulative_return_buy_hold_strategy = ((train_prices.ix[-1,:][0]- train_prices.ix[0,:][0])*100+sv)/sv - 1
print "cumulative return of buy-and-hold strategy on training:",\
        train_cumulative_return_buy_hold_strategy 

cumulative_return_buy_hold_strategy = ((prices.ix[-1,:][0]- prices.ix[0,:][0])*100+sv)/sv - 1
print "cumulative return of buy-and-hold strategy on testing:",\
        cumulative_return_buy_hold_strategy 

```

Turn off Q-learning and generate random actions:

```python
sv =10000
train_cumulative_return_buy_hold_strategy = ((train_prices.ix[-1,:][0]- train_prices.ix[0,:][0])*100+sv)/sv - 1
print "cumulative return of buy-and-hold strategy on training:",\
        train_cumulative_return_buy_hold_strategy 

cumulative_return_buy_hold_strategy = ((prices.ix[-1,:][0]- prices.ix[0,:][0])*100+sv)/sv - 1
print "cumulative return of buy-and-hold strategy on testing:",\
        cumulative_return_buy_hold_strategy 

# do num_simulatoins
num_simulation = 3

cumulative_returns_train = np.zeros(num_simulation)
cumulative_returns_test = np.zeros(num_simulation)

for i in range(0,num_simulation): 
    # instantiate the strategy learner
    # set learning to be False to generate random actions
    learner = sl.StrategyLearner(learning = False,verbose = False)

    # learning
    cumulative_return = learner.addEvidence(symbol = symbol,\
                sd = sd_train, \
                ed = ed_train, sv = 10000) 

    plt.plot(cumulative_return)

    #save the final result
    cumulative_returns_train[i] = cumulative_return[-1]
    print "cumulative_return of training:",cumulative_returns_train[i]

    # test the learner
    df_trades, cumulative_returns_test[i] = learner.testPolicy(symbol = symbol, sd = sd_test, \
    ed = ed_test, sv = 10000)

plt.title("Cumulative return on training set")
plt.ylabel("Cumulative return")
plt.xlabel("Trials")
plt.show()

print "Q-table:",learner.learner.Q[0:2200]
print "cumulative_returns_train", cumulative_returns_train 
print "cumulative_returns_test", cumulative_returns_test
```

Some observations:
- The graph shows cumulative returns are random and not dependent on the number of trials. 
- The cumulative returns on training and testing data make sense that if we trade randomly, the return might be negative. 
- Q-table are not updated and consist of zeros initially.
