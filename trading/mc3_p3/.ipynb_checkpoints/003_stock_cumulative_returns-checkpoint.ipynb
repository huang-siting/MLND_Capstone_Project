## Exploration of cumulative return of stocks

Plot the cumulative returns fot the three stocks in our dataset: IBM, SPY, GOOG.

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import QLearner as ql
import pandas as pd
import util as ut

# Pretty display for notebooks
%matplotlib inline
```

```python
stdate =dt.datetime(2004,8,19)
enddate =dt.datetime(2017,4,21) 

dates = pd.date_range(stdate, enddate)
syms=["SPY", "GOOG", "IBM"]
prices_all = ut.get_data(syms, dates)  
daily_cumulative_returns = ut.compute_cumulative_returns(prices_all)
```

```python
ut.plot_data(daily_cumulative_returns, title ="Cumulative returns", ylabel="cumulative returns")

```
